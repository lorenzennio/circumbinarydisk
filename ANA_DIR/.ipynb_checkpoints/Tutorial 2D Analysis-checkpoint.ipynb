# Analysis for 2D Athena++ HDF5 output

```python
import os, sys
import h5py
import numpy as np
import scipy as sp
import yt
import matplotlib.pyplot as plt
import matplotlib.colors
```

## Units
First we define some units for our system.
The below class is a copy uf the units.py file and can be imported for future use.

```python
class units:
    """
    Class to define units in cgs base and calculating som reference values
    Input:
        mbin: binary mass
        abin: binary separation (semi-major axis)
    """
    def __init__(self, mbin, abin):
        #cgs constants
        self.msol = 1.98847e33 #g
        self.mh   = 1.673534e-24 #g
        self.mu   = 2.3
        self.gn   = 6.6743e-8 #cm^3 g^-1 s^-2
        self.kb   = 1.3807e-16 #cm^2 g s^-2 K^-1
        self.au   = 1.495978707e13 #cm
        self.mbin    = mbin
        self.abin    = abin
        self.tbin    = np.sqrt((self.abin*self.au)**3/(self.mbin*self.msol*self.gn))
        self.densref = self.mbin*self.msol/(self.abin*self.au)**3
        self.presref = self.mbin*self.msol/(self.tbin**2 * self.abin*self.au) #g/(s^2 cm)
        self.tempref = self.mh * self.mu /self.kb*(self.mbin*self.msol*self.gn/(self.abin*self.au))
```

Next we specify units for our binary system:

```python
#Define the units
mbin = 3. #M_sun
abin = 50. #AU
u = units(mbin, abin)
```

## Load output files
We navigate to the directory where our output files are stored. They have the format 'cbd.out1.00000.athdf' etc.

```python
#directory to where your outputs are stored
os.chdir(os.path.abspath("__file__"+"/../../RUN_DIR/outputSMR2Dacc"))

#get all output files
prim = np.sort([f for f in os.listdir(os.getcwd()) if f.endswith(".athdf") and f.startswith("cbd.out1")])
print(prim)
```

Next we load the data into arrays. This procedure is performed in the "readdata.py" file.

```python
#take one example file
fprim = prim[-1]

#define a unit base for the data
unit_base= {"length_unit":(u.abin,"AU"), "time_unit":(u.tbin,"s"), "mass_unit":(u.mbin,"Msun")}


#load the data from the file
level=2 #this is the refinement level

ds_prim = yt.load(fprim, units_override = unit_base)

all_data_prim = ds_prim.covering_grid(level=level, \
                                    left_edge=ds_prim.domain_left_edge, \
                                    dims=ds_prim.domain_dimensions*2**level) 
                                    #dims = total dimensions 2**(refinement level)
#copy the data into arrays
rho = all_data_prim["rho"].in_units("g/cm**3").to_ndarray()[:,:,0]
#isothermal outputs do not contain pressure
try:
    press = all_data_prim["press"].in_units("g/(cm*s**2)").to_ndarray()[:,:,0]
except:
    pass
```

## Plot
Here the density is plotted as an example. You can just as easily plot the pressure or any other output. This is also performed in "readdata.py"

```python
frame = u.abin * 10.    #physical size of the frame
scale = 1.              #scale the plot in case you want to zoom in or out
lim = frame*scale       #limit the plot range
low=1e-15 #g/cm^3        lower cutoff
hig=1e-11 #g/cm^3        upper cutoff
lab=r"$\rho ~[g/cm^3]$" #label

cmap="plasma"           #color map
fs = 16                 #font size


fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

pos = ax.imshow(rho, cmap=cmap, origin='lower', extent=[-frame,frame,-frame,frame], \
                norm=matplotlib.colors.LogNorm(vmin=low, vmax=hig))

cbar = fig.colorbar(pos, ax=ax)

cbar.ax.tick_params(labelsize=fs)
cbar.set_label(lab, size=fs)

ax.set_aspect('equal', 'box')
ax.set_xlabel("$x_0~[AU]$", fontsize=fs)
ax.set_ylabel("$x_1~[AU]$", fontsize=fs)
ax.set_xlim(-lim, lim)
ax.set_ylim(-lim, lim)

for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(fs)
```
